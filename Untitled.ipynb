```python
class RandomKernel(object):
    def __init__(self, kernels, indices):
        self.len = kernels.shape[-1]         #Ex: 358
        self.indices = indices               #Ex: 0,1...,358
        np.random.shuffle(self.indices[0])   #shuffle normal (Ex: 294,558,...0,1,10)
        self.kernels = kernels[:, :, :, self.indices[0]] #kernels, mas em uma ordem aleatria
        self.index = 0
    
    def __iter__(self):
            return self

    def __next__(self):
        self.index =np.random.randint(len(self.indices[0]))
            
        n = self.kernels[:, :, :, self.index] #pega um kernel (o da fila - posicao index)
        self.index += 1
        return n #retorna um kernel
```

```python
import torch
import torch.nn as nn

torch.set_default_tensor_type(torch.DoubleTensor)

class SRMD(nn.Module):
    def __init__(self, num_blocks=11, num_channels=21, conv_dim=128, scale_factor=1):
        super(SRMD, self).__init__()
        self.num_channels = num_channels
        self.conv_dim = conv_dim
        self.sf = scale_factor

        self.nonlinear_mapping = self.make_layers(num_blocks)

        self.conv_last = nn.Sequential(
                            nn.Conv2d(self.conv_dim, 3*self.sf**2, kernel_size=3, padding=1),
                            nn.PixelShuffle(self.sf),
                            nn.Sigmoid()
                         )

    def forward(self, x):
        b_size = x.shape[0]
        h, w = list(x.shape[2:])
        x = self.nonlinear_mapping(x)
        x = self.conv_last(x)
        return x

    def make_layers(self, num_blocks):
        layers = []
        in_channels = self.num_channels
        for i in range(num_blocks):
            conv2d = nn.Conv2d(in_channels, self.conv_dim, kernel_size=3, padding=1)
            layers += [conv2d, nn.BatchNorm2d(self.conv_dim), nn.ReLU(inplace=True)]
            in_channels = self.conv_dim

        return nn.Sequential(*layers)
```

```python
def get_image(image_path, width, height):
    """
    Read image from image_path
    :param image_path: Path of image
    :param width: Width of image
    :param height: Height of image
    :return: Image data
    """
    
    image = Image.open(image_path)
    if np.array(image).shape==4:
        image = np.array(image)[:,:,:3]
        image = Image.fromarray(image)
    if image.size != (width, height):  # HACK - Check if image is from the CELEBA dataset
        # Remove most pixels that aren't part of a face
        face_width = face_height = 128
        j = (image.size[0] - face_width) // 2
        i = (image.size[1] - face_height) // 2
        image = image.crop([j, i, j + face_width, i + face_height])

    return image
```

```python
class SRMD(nn.Module):
    def __init__(self, num_blocks=11, num_channels=21, conv_dim=128, scale_factor=1):
        super(SRMD, self).__init__()
        self.num_channels = num_channels
        self.conv_dim = conv_dim
        self.sf = scale_factor

        self.nonlinear_mapping = self.make_layers(num_blocks)

        self.conv_last = nn.Sequential(
                            nn.Conv2d(self.conv_dim, 3*self.sf**2, kernel_size=3, padding=1),
                            nn.PixelShuffle(self.sf),
                            nn.Tanh()
                         )

    def forward(self, x):
        b_size = x.shape[0]
        h, w = list(x.shape[2:])
        x = self.nonlinear_mapping(x)
        x = self.conv_last(x)
        return x

    def make_layers(self, num_blocks):
        layers = []
        in_channels = self.num_channels
        for i in range(num_blocks):
            conv2d = nn.Conv2d(in_channels, self.conv_dim, kernel_size=3, padding=1)
            layers += [conv2d, nn.BatchNorm2d(self.conv_dim), nn.ReLU(inplace=True)]
            in_channels = self.conv_dim

        return nn.Sequential(*layers)
```
